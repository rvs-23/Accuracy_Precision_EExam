## Importing Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

from IPython.display import HTML

plt.style.use('fivethirtyeight')
np.random.seed(23)
```

## The Animation to visualize Accuracy and Precision...

### Drawing a sample figure on which to build the animation

```python
# Constructing 3 concentric circles

plt.scatter(0, 0, s= 20000, color='green', edgecolor='black', alpha=0.6, linewidth=1)
plt.scatter(0, 0, s= 10000, color='royalblue', edgecolor='black', alpha=0.6, linewidth=1)
plt.scatter(0, 0, s= 3000, color='red', edgecolor='black', alpha=0.75, linewidth=1.25)
plt.xlim(-0.5, 0.5)
plt.ylim(-0.5, 0.5)
plt.grid(False)
plt.title( "Scatter plot of points Circle" )
plt.show()
```

### Setting up the canvas

```python
%%time

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8.5,9))

x1, x2, x3, x4, y1, y2, y3, y4 = [], [], [], [], [], [], [], []
line1, = ax1.plot([], [], 'kX')
line2, = ax2.plot([], [], 'kX')
line3, = ax3.plot([], [], 'kX')
line4, = ax4.plot([], [], 'kX')


def init(color='royalblue'):
    """
    This function creates a canvas containing 3 concentric circles
    with 2 rows and 2 columns on which the animation will be built.
    One can also create another sub-function to create the plots.
    
    """
    ax1.set(xlim=(-0.6,0.6), ylim=(-0.6,0.6))
    ax1.set_title('High Precision Low Accuracy', fontsize=17)
    ax1.set_xticks([])
    ax1.scatter(0, 0, s= 20000, color=color, edgecolor='black', alpha=0.65, linewidth=1)
    ax1.scatter(0, 0, s= 10000, color='snow', edgecolor='black', alpha=0.8, linewidth=0.85)
    ax1.scatter(0, 0, s= 3000, color=color, edgecolor='black', alpha=0.75, linewidth=1.25)
#     ax1.spines["top"].set_visible(False)
#     ax1.spines["right"].set_visible(False)
#     ax1.spines["bottom"].set_visible(False)
    
    ax2.set(xlim=(-0.6,0.6), ylim=(-0.6,0.6))
    ax2.set_xticks([])
    ax2.set_title('High Accuracy Low Precision', fontsize=17)
    ax2.scatter(0, 0, s= 20000, color=color, edgecolor='black', alpha=0.65, linewidth=1)
    ax2.scatter(0, 0, s= 10000, color='snow', edgecolor='black', alpha=0.8, linewidth=0.85)
    ax2.scatter(0, 0, s= 3000, color=color, edgecolor='black', alpha=0.75, linewidth=1.25)
    
    ax3.set(xlim=(-0.6,0.6), ylim=(-0.6,0.6))
    ax3.set_xticks([])
    ax3.set_title('Low Precision Low Accuracy', fontsize=17)
    ax3.scatter(0, 0, s= 20000, color=color, edgecolor='black', alpha=0.65, linewidth=1)
    ax3.scatter(0, 0, s= 10000, color='snow', edgecolor='black', alpha=0.8, linewidth=0.85)
    ax3.scatter(0, 0, s= 3000, color=color, edgecolor='black', alpha=0.75, linewidth=1.25)
    
    ax4.set(xlim=(-0.6,0.6), ylim=(-0.6,0.6))
    ax4.set_xticks([])
    ax4.set_title('High Accuracy High Precision', fontsize=17)
    ax4.scatter(0, 0, s= 20000, color=color, edgecolor='black', alpha=0.65, linewidth=1)
    ax4.scatter(0, 0, s= 10000, color='snow', edgecolor='black', alpha=0.8, linewidth=0.85)
    ax4.scatter(0, 0, s= 3000, color=color, edgecolor='black', alpha=0.75, linewidth=1.25)
    
    
    return line1, line2, line3, line4,
plt.close()
```

### Animation

```python
%%time

# List of points for the high precision and low accuracy board
prec_x = [0.23, 0.24, 0.28, 0.26, 0.27, 0.29]
prec_y = [0.2, 0.18, 0.17, 0.215, 0.205, 0.185]

def animate(frame):
    if frame%19==0:                                       # plotting every 19th frame
        frame = frame//19
        x1.append(prec_x[frame])
        y1.append(prec_y[frame])
        line1.set_data(x1, y1)

        x2.append(round(np.random.uniform(-0.09, 0.09), 3))
        y2.append(round(np.random.uniform(-0.09, 0.09), 3))
        line2.set_data(x2, y2)

        x3.append(round(np.random.uniform(-0.3, 0.3), 3))
        y3.append(round(np.random.uniform(-0.3, 0.3), 3))
        line3.set_data(x3, y3)

        x4.append(round(np.random.uniform(-0.05, 0.05), 3))
        y4.append(prec_x[frame]-0.25)
        line4.set_data(x4, y4)

    return line1, line2, line3, line4,

ani = FuncAnimation(fig, animate, frames=110, init_func=init, interval=60, blit=True)
plt.close()
HTML(ani.to_html5_video())
```
